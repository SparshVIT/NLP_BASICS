# NLP BASICS 

**Suppose we have a text based data, then we have to perform the NLP processing, 
The First Process is Tokenization, here we provide the paragraph or sentence, then we convert it to sentence or words through the tokenization**

**1. Import the library**

```python
import nltk
```

**2. Download all the packages in nltk library**

```python
nltk.download()
```

```python
paragraph = """Mumbai and Bangalore are two of the most prominent cities in India when it comes to software engineering opportunities. Both cities have a thriving IT industry, with several big tech companies having their offices in these cities. Mumbai is considered as the financial capital of India, while Bangalore is known as the Silicon Valley of India.
In terms of software engineering opportunities, Bangalore has an edge over Mumbai. Bangalore has a larger concentration of IT companies, which means there are more job opportunities for software engineers. Bangalore is home to some of the biggest IT giants in the world such as Infosys, Wipro, TCS, and IBM. The city is also home to several startups, which means there are opportunities for software engineers to work with new technologies and innovative ideas.
However, Mumbai is not far behind when it comes to software engineering opportunities. The city is home to several global IT companies such as Capgemini, Accenture, and JP Morgan. Additionally, Mumbai has a growing startup ecosystem with companies such as Ola, Paytm, and Swiggy having their offices in the city. Mumbai also has a thriving fintech industry, which means there are opportunities for software engineers to work with financial technologies.
When it comes to life, Mumbai and Bangalore have their own unique advantages and disadvantages. Mumbai is a bustling metropolis, with a diverse culture and a vibrant nightlife. The city has several tourist attractions such as the Gateway of India, Marine Drive, and the Elephanta Caves. However, Mumbai is also known for its high cost of living, traffic congestion, and pollution.
On the other hand, Bangalore is a relatively quieter city, with a pleasant climate and a laid-back lifestyle. The city has several parks and lakes, which provide a much-needed respite from the hustle and bustle of the city. Bangalore is also known for its rich cultural heritage, with several ancient temples and historical monuments such as the Bangalore Palace and the Bull Temple. However, the city's rapid growth in recent years has led to traffic congestion and pollution, which can be a downside for some.
In conclusion, both Mumbai and Bangalore have their own advantages and disadvantages when it comes to software engineering opportunities and life. While Bangalore has a larger concentration of IT companies and a more laid-back lifestyle, Mumbai has a growing startup ecosystem and a vibrant culture. Ultimately, the choice between Mumbai and Bangalore will depend on the individual's preferences and priorities."""
```

**To tokenize the paragraph to sentence we will use
nltk.sent_tokenize function, here it will apply regex and convert this paragraph into bunch of sentences**

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

**If we want to convert this paragraph into bunch of words  then we use nltk.word_tokenize(paragraph)**

```python
words = nltk.word_tokenize(paragraph)
```

```python
words
```
